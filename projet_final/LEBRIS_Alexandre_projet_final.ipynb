### Scrapping des donn√©es sur les bateaux

On "scrape" au sein d'un DataFrame les donn√©es sur les bateaux, pr√©sent√©s sur le site du Vend√©e Globe ((si le fichier final `\data\df_coureur.pickle` n'existe pas encore) :

```python
def download_boat_data():
    # Chargement des donn√©es sur les bateaux
    page = requests.get(PATH_RANK_BOAT)
    soup = BeautifulSoup(page.content, 'html.parser')
    boat_list = soup.select('div.boats-list__popup')
    
    # Extraction des informations au sein d'un Dataframe
    data = []
    for boat in boat_list:
        specs = boat.select('li')
        boat_data = {
            key: value for key, value in
            [spec.contents[0].split(" : ")[:2] for spec in specs]
        }
        data.append(boat_data)
        
    return pd.DataFrame(data=data)
```

### D√©finition de fonctions utiles

On d√©finit dans un premier temps des fonctions *ad hoc* pour la r√©paration des fichiers Excel et la conversion des variables :

```python
# Fonction pour "r√©parer" les fichiers Excel t√©l√©charg√©s (pr√©sence 
# d'un mot-cl√© qui emp√™che l'ouverture du fichier par openpyxl et donc pandas)
def fix_excel(file_name, name_filter, change):
    tempdir = tempfile.mkdtemp()
    
    try:
        tempname = os.path.join(tempdir, 'new.zip')
        with ZipFile(file_name, 'r') as r, ZipFile(tempname, 'w') as w:
            for item in r.infolist():
                data = r.read(item.filename)           
                data = change(data)
                w.writestr(item, data)
        shutil.move(tempname, file_name)
    finally:
        shutil.rmtree(tempdir)
```

```python
# Fonction pour convertir les degr√©s, minutes, secondes en d√©gr√©s d√©cimaux
def dms2dd(dms):
    degree, minute, second = re.findall("(\d+)", dms)
    dd = int(degree) + int(minute) / 60 + int(second) / 3600
    
    if dms[-1] in 'SW':
        return -dd
    return dd
```

```python
# Fonction pour convertir les √©carts en temps en timedelta
def duration2timedelta(item):
    if item is np.nan:
        return item
    
    figures = re.findall("(\d+)", item)
    
    if len(figures) == 4:
        days, hours, minutes, seconds = figures
    elif len(figures) == 3:
        hours, minutes, seconds = figures
        days = 0
    else:
        minutes, seconds = figures
        days= 0
        hours = 0
        
    return datetime.timedelta(days=int(days), seconds=int(seconds),
                              minutes=int(minutes), hours=int(hours))
```

```python
# Fonction pour convertir en valeur num√©riqueune cha√Æne de caract√®re codant une longueur
def length2num(string):
    try:
        res = float(re.findall("([\d,]+)", string)[0].replace(',', '.'))
    except IndexError:
        res = np.nan
    return res
```

```python
# Fonction pour convertir en valeur num√©rique une cha√Æne de caract√®re codant une surface
def area2num(string):
    return int(re.findall("(\d+)", string)[0])
```

### Chargement et nettoyage des donn√©es de course

```python
# Fonction pour charger et nettoyer les donn√©es d'un fichier Excel
def import_clean_excel(filename):   
    columns = ["rang", "voile", "nom_skipper", "date_heure", "lat", "long",
               "cap_30min", "vit_30min", "vmg_30min", "distance_30min",
               "cap_prec", "vit_prec", "vmg_prec", "distance_prec",
               "cap_24h", "vit_24h", "vmg_24h", "distance_24h",
               "dtf", "dtl"]

    # Import du fichier Excel dans un DataFrame
    df = pd.read_excel(filename, skiprows=4, nrows=33, usecols="B:U")
    df.columns = columns

    # Extraction des lignes pour les coureurs toujours en course
    df = df[(~df.nom_skipper.isna()) & (~df.date_heure.isna())]
    
    if len(df) == 0:
        return df

    # Traitement de la colonne rang
    df.rang = pd.to_numeric(df.rang)

    # Traitement de la colonne voile
    df.voile = df.voile.apply(lambda line: re.findall("(\d+)", line)[0])

    # Traitement de la colonne nom_skipper
    df.nom_skipper = (df.nom_skipper
                         .apply(lambda line: re.match("^(.*)\\n", line).group(1))
                         .str.lower())

    # Traitement de la colonne heure
    date = filename[17:25]
    heure = filename[26:32]
    df.date_heure = df.date_heure.apply(lambda line: date + " " + heure)
    df.date_heure = pd.to_datetime(df.date_heure, format="%Y%m%d %H%M%S")
    df.date_heure = df.date_heure - datetime.timedelta(minutes=30)

    # Traitement des colonnes latitude et longitude
    df["geo"] = df.apply(lambda line: (dms2dd(line.lat), dms2dd(line.long)), axis=1)

    # Traitement des colonnes heading
    columns = ["cap_30min", "cap_24h", "cap_prec"]
    df[columns] = df[columns].applymap(lambda item: float(item[:-1]))

    # Traitement des colonnes speed
    columns = ["vit_30min", "vit_24h", "vit_prec"]
    df[columns] = df[columns].applymap(lambda item: float(item[:-4]))

    # Traitement des colonnes vmg
    columns = ["vmg_30min", "vmg_24h", "vmg_prec"]
    df[columns] = df[columns].applymap(lambda item: float(item[:-4]))

    # Traitement des colonnes distance
    columns = ["distance_30min", "distance_24h", "distance_prec"]
    df[columns] = df[columns].applymap(lambda item: float(item[:-3]))

    # Traitement des colonnes dtf et dtl
    columns = ["dtf", "dtl"]
    df[columns] = df[columns].applymap(lambda item: float(item[:-3]))
    
    return df
```

```python
# Aggr√©gation des donn√©es dispers√©es sur les diff√©rents fichiers Excel au sein d'un unique Dataframe,
# en utilisant la fonction d√©finie ci-dessus
def generate_df_course(filename_list):   
    df_list = []
    
    # R√©paration, chargement et nettoyage de chaque fichier Excel
    for filename in filename_list:
        fix_excel(filename, 
                  name_filter='xl/styles.xml', # L'anomalie se trouve dans le fichier xml styles
                  change=lambda d: re.sub(b'xxid="\d*"', b"", d)) # On enl√®ve le mot-cl√© xxid
        df_list.append(import_clean_excel(filename))
    
    # Cr√©ation du Dataframe final
    df_course = pd.concat(df_list)
    
    # Suppression des dupliqu√©s
    df_course = df_course.drop_duplicates()
    
    # Suppression des colonnes inutiles et reset
    df_course = df_course.drop(columns=['lat', 'long'])
    df_course = df_course.reset_index().drop(columns='index')
    
    return df_course
```

```python
if not os.path.isfile("data/df_course.pickle"):
    df_course = generate_df_course(filename_list)
```

### Chargement et nettoyage des donn√©es sur les coureurs

```python
# Cr√©ation du DataFrame avec le tableau final et les donn√©es d√©taill√©es 
# sur les coureurs
def generate_df_coureur(filename_list):   
    # Import du dernier fichier Excel en date (avec tous les r√©sultats finaux)
    df_coureur = pd.read_excel(sorted(filename_list)[-1],
                               skiprows=4, nrows=33, usecols="B:D, H:I, N, P:U")
    columns = ["status", "nat_sail", "skipper_crew", "date_arrivee", "temps_course",
               "ecart_premier", "ecart_prec", "vit_ortho", "distance_ortho",
               "arrow", "vit_fond", "distance_fond"]
    df_coureur.columns = columns

    # Cr√©ation de la colonne final_ranking
    df_coureur['final_ranking'] = df_coureur.status.apply(lambda item: re.findall("(\d*)", item)[0])
    df_coureur.final_ranking = pd.to_numeric(df_coureur.final_ranking)

    # Modification de la colonne status
    df_coureur.status = df_coureur.status.apply(lambda item: "ARV" if "ARV" in item else "RET")

    # Cr√©ation des colonnes skipper et crew
    df_coureur['nom_skipper'] = (df_coureur.skipper_crew
                                 .apply(lambda item: re.findall("^(.*)\\n", item)[0]))
    df_coureur['nom_bateau'] = (df_coureur.skipper_crew
                              .apply(lambda item: re.findall("\\n(.*)", item)[0]))

    # Cr√©ation des colonnes nat et sail
    df_coureur['nat'] = df_coureur.nat_sail.apply(lambda item: re.findall("\\n(\w*)", item)[0])
    df_coureur['voile'] = df_coureur.nat_sail.apply(lambda item: re.findall("(\d*)$", item)[0])

    # Nettoyage de la colonne date_arrivee
    df_coureur.date_arrivee = pd.to_datetime(df_coureur.date_arrivee, format="%d/%m/%Y %H:%M:%S FR")

    # Nettoyage de la colonne temps_course
    df_coureur.temps_course = df_coureur.temps_course.apply(lambda item: item.split("\n")[0]
                                                      if isinstance(item, str) else item)
    df_coureur.temps_course = df_coureur.temps_course.apply(duration2timedelta)

    # Cr√©ation de la colonne compensation
    compensation = {
        "17": datetime.timedelta(hours=-10, minutes=-15),
        "01": datetime.timedelta(hours=-16, minutes=-15),
        "10": datetime.timedelta(hours=-6),
        "02": datetime.timedelta(hours=2)
    }
    df_coureur['compensation'] = df_coureur['voile'].map(compensation)

    # Conversion des colonnes ecart_premier et ecart_prec
    df_coureur.ecart_premier = df_coureur.ecart_premier.apply(duration2timedelta)
    df_coureur.ecart_prec = df_coureur.ecart_prec.apply(duration2timedelta)

    # Conversion des colonnes speed
    columns = ["vit_ortho", "vit_fond"]
    df_coureur[columns] = df_coureur[columns].applymap(lambda item: float(item[:-4])
                                                       if item is not np.nan else item)

    # Conversion des colonnes distance
    columns = ["distance_ortho", "distance_fond"]
    df_coureur[columns] = df_coureur[columns].applymap(lambda item: float(item[:-3])
                                                       if item is not np.nan else item)

    # Conversion de la colonne arrow
    df_coureur.arrow = df_coureur.arrow.apply(lambda item: float(item[:-1])
                                              if item is not np.nan else item)

    # Suppression des colonnes inutiles
    columns = ['skipper_crew', "nat_sail"]
    df_coureur = df_coureur.drop(columns=columns)
    
    return df_coureur
```

### Nettoyage des donn√©es sur les bateaux

```python
# Fonction qui nettoie les donn√©es scrapp√©es relatives aux caract√©ristiques des diff√©rents bateaux
def clean_df_bateau(df_bateau):  
    # Extraction de la date de lancement
    df_bateau['Date de lancement'] = (df_bateau['Date de lancement']
                              .apply(lambda date: int(re.findall("(\d{4})", date)[0])))

    # Conversion des mesures de longeur, de poids ou de surface en num√©rique
    columns = ['Longueur', 'Largeur', "Tirant d'eau", 'Hauteur m√¢t',
               'D√©placement (poids)']
    df_bateau[columns] = df_bateau[columns].applymap(length2num)

    columns = ['Surface de voiles au pr√®s', 'Surface de voiles au portant']
    df_bateau[columns] = df_bateau[columns].applymap(area2num)

    # Nettoyage de la colonne Nombre de d√©rives
    df_bateau['Nombre de d√©rives'] = (df_bateau['Nombre de d√©rives']
                                      .replace({'2 asym√©triques': '2',
                                                'foiler': 'foils'}))

    # Nettoyage de la colonne Voile quille
    df_bateau['Voile quille'] = (df_bateau['Voile quille']
                                      .replace({'Acier m√©cano soud√©': 'acier m√©cano soud√©',
                                                'basculante sur v√©rin hydraulique': 'basculante avec v√©rin',
                                                'Inox usin√©': 'inox usin√©'}))

    # Extraction du num√©ro de voile et nettoyage du num√©ro
    df_bateau["Num√©ro de voile"].replace('16', 'MON 10', inplace=True)
    df_bateau['Num√©ro de voile'] = df_bateau['Num√©ro de voile'].fillna('59')

    df_bateau['Num√©ro de voile'] = (df_bateau['Num√©ro de voile']
                                    .apply(lambda line: str(line)))
    df_bateau["voile"] = (df_bateau['Num√©ro de voile']
                          .apply(lambda line: re.findall("(\d+)", line)[0]))
    df_bateau.voile = (df_bateau.voile
                       .replace({"08": "8", "77": "777",
                                 "001": "01", "07": "7", "2": "02"}))

    # Renommage des colonnes
    col2col = {
        'Anciens noms du bateau': 'ancien_noms_bateau',
        'Architecte': 'architecte',
        'Chantier': 'chantier',
        'Date de lancement': 'annee_lacement',
        'Longueur': 'longueur',
        'Largeur': 'largeur',
        "Tirant d'eau": 'tirant_eau',
        'D√©placement (poids)': 'poids',
        'Nombre de d√©rives': 'derives',
        'Hauteur m√¢t': 'hauteur_mat',
        'Voile quille': 'quille',
        'Surface de voiles au pr√®s': 'surface_voiles_pres',
        'Surface de voiles au portant': 'surface_voiles_portant'
    }
    df_bateau = df_bateau.rename(columns=col2col)
    
    # Nouvelles valeurs pour le poids non renseign√© de 2 bateaux par la moyenne
    df_bateau.poids = df_bateau.poids.fillna(df_bateau.poids.mean())

    # Suppression des colonnes inutiles
    df_bateau = df_bateau.drop(columns=['Num√©ro de voile'])
    
    return df_bateau
```

```python
# Calcul de la zone g√©ographique o√π sont localis√©e les bateaux
def generate_zones(df_course):
    zones = np.zeros(len(df_course))
    
    for i in df_course.idx_skipper.unique():
        foo = df_course[df_course.idx_skipper == i].sort_values(by='date_heure')
        idx = foo.geo.apply(lambda point: point[0]>0).argmin()
        if idx == 0:
            zones[foo.geo.index] = 0
            continue
        zone0 = list(foo.geo.iloc[:idx].index)
        zones[zone0] = 0

        foo = foo[idx:]
        idx = foo.geo.apply(lambda point: point[1]>18.4739).argmax()
        if idx == 0:
            zones[foo.geo.index] = 1
            continue
        zone1 = list(foo.geo.iloc[:idx].index)
        zones[zone1] = 1

        foo = foo[idx:]
        idx = foo.geo.apply(lambda point: point[1]>115.1358).argmax()
        if idx == 0:
            zones[foo.geo.index] = 2
            continue
        zone2 = list(foo.geo.iloc[:idx].index)
        zones[zone2] = 2

        foo = foo[idx:]
        idx = foo.geo.apply(lambda point: np.abs(point[1])<67.2892).argmax()
        if idx == 0:
            zones[foo.geo.index] = 3
            continue
        zone3 = list(foo.geo.iloc[:idx].index)
        zones[zone3] = 3

        foo = foo[idx:]
        idx = foo.geo.apply(lambda point: point[0]>0).argmax()
        if idx == 0:
            zones[foo.geo.index] = 4
            continue
        zone4 = list(foo.geo.iloc[:idx].index)
        zones[zone4] = 4


        foo = foo[idx:]
        zone5 = list(foo.index)
        zones[zone5] = 5
        
    df_course['zones'] = zones
    return df_course
```
